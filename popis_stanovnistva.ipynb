```python
import pandas as pd
```

```python
data_2021 = pd.read_excel("Popis_2021-Stanovnistvo_po_gradovima_opcinama.xlsx",sheet_name=3,header=7)
data_2011 = pd.read_excel("Popis_vjera_2011.xls",sheet_name=0,header=1)
```

```python
del data_2021['County of'], data_2021['Local self-government unit']
data_2021
```

```python
data_2011.rename(columns={'Ime županije':'Županija','Ime grada/općine':'Grad/općina','Grad ili općina':'Jedinica lokalne samouprave'},inplace=True)
data_2011
```
